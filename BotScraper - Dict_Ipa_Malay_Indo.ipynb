```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import pandas as pd
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup as bs
```

```python
browser = webdriver.Chrome(executable_path=r"C:\Users\Hanif\Anaconda Projek\Botika Internship\Botika\Project\Sub Team TTS\ipaBotScrapper\chromedriver.exe")
```

```python
df = pd.read_csv('kata.csv')
```

```python
df
```

```python
ipa = []
```

```python
for kata in df['katakunci']:
    browser.get('https://open-dict-data.github.io/ipa-lookup/ma/?#')
    text_field = browser.find_element_by_id("input")
    text_field.clear()
    text_field.send_keys(kata)
    search = browser.find_element_by_class_name('btn')
    search.click()
    ipa.append(browser.find_element_by_id("output").text)
```

```python
ipa
```

```python
df['ipa'] = ipa
```

```python
df
```

```python
def cleansing(x):
    if '\n' in x:
        x = x.split('\n')
        return x[1]
    elif 'not found...' in x:
        return ' '
```

```python
df['ipa'] = df['ipa'].apply(lambda x : cleansing(x))
```
